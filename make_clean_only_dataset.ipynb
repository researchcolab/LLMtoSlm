# Make Clean-Only Dataset
Filters `training_data_augmented_v2.jsonl` to keep only the 446 clean (no-backtrack) examples.

**Run this first** before training the clean-only ablation model.

## Configuration
Edit the paths below if your files are in a different location.

```python
INPUT_PATH  = "training_data_augmented_v2.jsonl"  # full augmented dataset (1487 rows)
OUTPUT_PATH = "training_data_clean_only.jsonl"     # output: 446 clean rows only
```

## Filter and write

```python
import json

with open(INPUT_PATH) as f:
    all_rows = [json.loads(line) for line in f if line.strip()]

clean_rows = [r for r in all_rows if r["example_type"] == "clean"]

with open(OUTPUT_PATH, "w") as f:
    for row in clean_rows:
        f.write(json.dumps(row) + "\n")

print(f"Total rows read : {len(all_rows)}")
print(f"Clean rows kept : {len(clean_rows)}")
assert len(clean_rows) == 446, f"Expected 446 clean rows, got {len(clean_rows)}"
print("Count matches expected 446 ✓")
print(f"Written to      : {OUTPUT_PATH}")
```
